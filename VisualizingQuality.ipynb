```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

"""
This is a python script that analyse the Amplicon_Quality.csv and output a pie chart
"""

df = pd.read_csv('Amplicon_Quality.csv')
comma = df.iloc[2:,2].astype(str).astype(float).sum()
seven = df.iloc[2:,3].astype(str).astype(float).sum()
colon = df.iloc[2:,4].astype(str).astype(float).sum()
E = df.iloc[2:,5].astype(str).astype(float).sum()
F = df.iloc[2:,6].astype(str).astype(float).sum()
I = df.iloc[2:,7].astype(str).astype(float).sum()
```

```python
#define data
data = [comma, seven, colon, E, F, I]
labels = ['11', '22', '25', '36', '37', '40']

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:6]

#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
pie, ax = plt.subplots(figsize=[10,6])
plt.pie(x=data, autopct="%.1f%%", explode=[0.05]*6, labels=labels, pctdistance=0.5)
plt.title("Phred scores", fontsize=14);
```
