```python
import pandas as pd

lib = pd.read_csv ('CheckMutatioon_Library.csv')
unsorted = pd.read_csv ('CheckMutatioon_Unsorted.csv')
ap1 = pd.read_csv ('CheckMutatioon_AP1.csv')
ap2 = pd.read_csv ('CheckMutatioon_AP2.csv')
ap3 = pd.read_csv ('CheckMutatioon_AP3.csv')
```

```python
lib['AA'] = lib['Position'] + ":" + lib['AminoAcid']
lib = lib.drop(columns=['Position', 'AminoAcid', 'Unnamed: 0'])
unsorted['AA'] = unsorted['Position'] + ":" + unsorted['AminoAcid']
unsorted = unsorted.drop(columns=['Position', 'AminoAcid', 'Unnamed: 0'])
ap1['AA'] = ap1['Position'] + ":" + ap1['AminoAcid']
ap1 = ap1.drop(columns=['Position', 'AminoAcid', 'Unnamed: 0'])
ap2['AA'] = ap2['Position'] + ":" + ap2['AminoAcid']
ap2 = ap2.drop(columns=['Position', 'AminoAcid', 'Unnamed: 0'])
ap3['AA'] = ap3['Position'] + ":" + ap3['AminoAcid']
ap3 = ap3.drop(columns=['Position', 'AminoAcid', 'Unnamed: 0'])
```

```python
combine = pd.merge(lib,unsorted)
combine = pd.merge(combine,ap1)
combine = pd.merge(combine,ap2)
combine = pd.merge(combine,ap3)
combine
```

```python
combine.to_csv('combined_data.csv')
```
