```python
import numpy as np 
import seaborn as sns
import pandas as pd
import warnings
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
import math
warnings.filterwarnings('ignore')
sns.set_theme()

df = pd.read_csv("combined_data.csv")
```

```python
ap1 = df[['AA','AP1_Percent']]
ap1 = ap1.rename(columns={"AP1_Percent":'Percent'})
ap1['Cell'] = 'AP1'
ap2 = df[['AA','AP2_Percent']]
ap2 = ap2.rename(columns={"AP2_Percent":'Percent'})
ap2['Cell'] = 'AP2'
ap3 = df[['AA','AP3_Percent']]
ap3 = ap3.rename(columns={"AP3_Percent":'Percent'})
ap3['Cell'] = 'AP3'
lib = df[['AA','Lib_Percent']]
lib = lib.rename(columns={"Lib_Percent":'Percent'})
lib['Cell'] = 'Library'
unsorted = df[['AA','Unsorted_Percent']]
unsorted = unsorted.rename(columns={"Unsorted_Percent":'Percent'})
unsorted['Cell'] = 'Unsorted'
```

```python
frames = [ap1,ap2,ap3,lib,unsorted]
newdf = pd.concat(frames, ignore_index=True)
newdf
```

```python
sns.displot(newdf, x="Percent", hue="Cell", kind="kde", fill=True, log_scale = True)
```
